```python
X = data.drop(columns = "Outcome", axis=1)
y = data["Outcome"]
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaler.fit(X)
scaler_features = scaler.transform(X)
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test,y_train, y_test = train_test_split(scaler_features,y, test_size=0.2, random_state=42)
```

Tensorflow estimator ValueError: logits and labels must have the same shape ((?, 1) vs (?,))

If you're doing binary cross-entropy, then your dataset probably has 2 classes and the error is coming because your label vectors (both in testing and training) have the form [0,1,0,1,1,1,0,0,1,...].

and it need  [[1. 0.],[0. 1.],[1. 0.],[0. 1.]...]

To solve the problem we use one hot encoing

```python
from tensorflow.keras.utils import to_categorical
print(to_categorical(y_train))
print("\n")
print(y_train)
```

```python
x_train[:5]
```

```python
y_train = to_categorical(y_train)
```

```python
from tensorflow.keras.layers import Dense
from tensorflow.keras.models import Sequential
```

```python
model = Sequential()
model.add(Dense(12, activation="relu", input_shape=(8,)))
model.add(Dense(8, activation="relu"))
model.add(Dense(2, activation="sigmoid"))
```

```python
model.compile(loss="binary_crossentropy",optimizer="adam",metrics=['accuracy'])
```

```python
model.fit(x_train,y_train,epochs=150,batch_size=10)
```

```python
from sklearn.metrics import accuracy_score
y_pred = model.predict(x_test)
print(y_pred)
```

```python
y_pred = np.argmax(y_pred,axis=1)
print(y_pred)
```

```python
accuracy_score(y_test,y_pred)
```
